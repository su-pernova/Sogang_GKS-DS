##**4) Scatterplots**

```
from matplotlib import pyplot as plt
scores = [70, 60, 90, 80, 60, 55, 50]
minutes = [150, 100, 300, 350, 230, 200, 50]
labels = ['chun-fung', 'wing-yee', 'marlen', 'jian', 'minki', 'eojin','sumi']
assert len(scores)==len(minutes)==len(labels), "paired data must be same lengths"

plt.scatter(scores, minutes)
for label, score, time in zip(labels, scores, minutes):
  plt.annotate(label, xy=(score, time), xytext=(5,-5), textcoords='offset points')

plt.title('Daily minutes vs Mid-exam Scores')
plt.xlabel('Scores')
plt.ylabel('daily minutes spend on the course')
plt.show()
```

```
import json
import requests

def get_page_data(url):
  data = requests.get(url)
  repos = json.loads(data.text)
  print("cur page:", url[-6:])
  page_data = [ ]
  for i, each in enumerate(repos):
    if each != 'storeInfos':
      print(i,each,repos[each])
    else : 
      first_store=lambda dic_list:(dic_list[0])
      if len(repos[each])>0:
        print(i,each,first_store(repos[each]))
        page_data = repos[each]
  return page_data
```

```
region_list=['경기', '광주', '대구', '대전', '부산', '서울', '제주','그외']
region_stores_lists = {each:list() for each in region_list}
print('init lists', region_stores_lists)

for region, store_list in region_stores_lists.items():
  for each in all_list:
    if each['addr'].startswith(region) : store_list.append((each['addr'], each['name']))

l_list=[]
for each in all_list:
  include = 0
  for region in region_list:
    if each['addr'].startswith(region) : include = 1
  if include == 0:
    l_list.append((each['addr'], each['name']))

region_stores_lists['그외'] = l_list
for r in region_stores_lists:print(r,region_stores_lists[r][0:2])
```

**Draw a pie chart with labels and legend**

```
import matplotlib.pyplot as plt
import numpy as np
fig, ax = plt.subplots(figsize=(6,3), subplot_kw=dict(aspect="equal"))

recipe = ["375 g flour", "75 g sugar", "250 g butter", "300 g berries"]
data=[float(x.split()[0]) for x in recipe]
ingredients = [x.split()[-1] for x in recipe]

def func(pct, allvals):
  absolute = int(pct/100.*np.sum(allvals))
  return"{:.1f}%\n({:d} g)".format(pct,absolute)

wedges, texts, autotexts = ax.pie(data, autopct=lambda pct: func(pct, data), textprops=dict(color='w'))
ax.legend(wedges, ingredients, title="Ingredients", loc="center left", bbox_to_anchor=(1,0,0.5,1))
plt.setp(autotexts, size=8, weight="bold")
ax.set_title("Matplotlib bakery : A pie")
plt.show()
```
